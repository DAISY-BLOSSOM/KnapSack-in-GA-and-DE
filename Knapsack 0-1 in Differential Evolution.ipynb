```python
import numpy as np
import recordclass as cols
import math 
import random as rnd
import copy as cpy

################# UTIL FUNCTIONS ########################
def arrayCopy(arr1,arr2) : 
    for i in range (len(arr1)):
        arr2.append(arr1[i])
    return arr2
def differenceOfTwoVectors (obj1,obj2) :
    temp = cpy.deepcopy(obj1)
    for i in range(len(temp.vector)):
        temp.vector[i] = max(0,obj1.vector[i]-obj2.vector[i]) 

    return temp

def sumOfTwoVectors (obj1,obj2,itemsArray,maxBagWeight) :
    temp = cpy.deepcopy(obj1)
    currentTotalWeight = 0;
    for i in range(len(temp.vector)):
        currentItemsFrequency = temp.vector[i] + obj2.vector[i]
        if(currentItemsFrequency * itemsArray[i].weight + currentTotalWeight > maxBagWeight):
            temp.vector[i] = 0
        else :
            temp.vector[i] += obj2.vector[i]
            currentTotalWeight += currentItemsFrequency * itemsArray[i].weight; 
        
    return temp
    
    
def updateWeight(obj,itemsArray):
    obj.weight = 0;
    for i in range(len(obj.vector)):
        obj.weight+= (obj.vector[i] * itemsArray[i].weight)
            
    return

def weightBy(temp,weight):
    for i in range(len(temp.vector)):
        temp.vector[i] = math.ceil(temp.vector[i] * weight)
    return temp
def getValOfVector(vector,itemsArray):
    sum = 0
    for i in range (len(vector)):
        sum += vector[i] * itemsArray[i].value
    return sum
    

  
    #########################################

Item = cols.recordclass('item', 'value weight')
VectorWrapper = cols.recordclass('VectorWrapper', 'vector weight')


####crossover and selection occurs here
def crossOver(populationArray,mutatedArray,crR,maxBagWeight):
    temp = []
    lenOfPopArray = len(populationArray)
    
    while len(temp) < lenOfPopArray :
        offSpringVector = []
      
        if (rnd.uniform(0,1) < crR ):
            offSpringVector = generateRandomVectors(populationArray,1)
        else :        
            offSpringVector = generateRandomVectors(mutatedArray,1)
            
        temp.append(offSpringVector)
        
        
    return temp


def generateRandomVectors(tempArray,noOfVectors,evaluateMaxWeight = None) :
    res = []
    for i in range(noOfVectors):
        if (len(tempArray) == 0):
            break
        randomIndex = rnd.randint(0,len(tempArray)-1)
        randomVector = tempArray[randomIndex]
        res.append(randomVector)
        tempArray.pop(randomIndex)
      
    if(noOfVectors == 1):
        return res[0]
    return res

def generateMutatedArray(populationArray,itemsArray,weight,maxBagWeight):
    temp=[]    
    mutatedArray = []
    for pivot in range(len(populationArray)) :
        temp = arrayCopy(populationArray,temp)
        temp.pop(pivot)
        randomGeneratedVectors = generateRandomVectors(temp,3)
        generatedVector = sumOfTwoVectors(weightBy(differenceOfTwoVectors(randomGeneratedVectors[0],randomGeneratedVectors[1]),weight),randomGeneratedVectors[2],itemsArray,maxBagWeight)
        updateWeight(generatedVector,itemsArray)
        mutatedArray.append(generatedVector)
        
    return mutatedArray


def generateVector(itemsArray,noOfItems,maxBagWeight) : 

    temp = []
    for i in range(noOfItems):
        maxNumber = int(maxBagWeight / itemsArray[i].weight)
        randomNumber = rnd.randint(0,math.floor(maxNumber))
        maxBagWeight-=randomNumber * itemsArray[i].weight
        temp.append(randomNumber)  
        
    res = VectorWrapper(temp,0)   
    updateWeight(res,itemsArray)
    return res


def generatePopulation (populationSize , itemsArray , noOfItems,maxBagWeight) :
    populationArray = []
    for i in range (populationSize) :
        populationArray.append(generateVector(itemsArray,noOfItems,maxBagWeight)) 
        
    return populationArray

def generateItems(maxValue, maxWeight , noOfItems): 
    temp = []
    for i in range(noOfItems):
        randomMaxWeight = (rnd.uniform(1, maxWeight) * 0.5) + 1
        item = Item(rnd.uniform(1,maxValue) , randomMaxWeight  )
        maxWeight-=randomMaxWeight
        temp.append(item)
        
    return temp

"""
def generateItems(maxValue, maxWeight , noOfItems): 
    temp = []
    arr = [[6,5],[5,6],[6,4],[6,6],[3,5],[7,2]]
    for i in range(len(arr)):
        item = Item(arr[i][0] ,arr[i][1])
        temp.append(item)
        
    return temp
"""


generations = 50                   
maxBagWeight = 15
maxItemValue = 100
noOfItems = 6
populationSize = 10
weight = 0.5
crR= 0.6



itemsArray = generateItems(maxItemValue,maxBagWeight,noOfItems)
populationArray = generatePopulation(populationSize,itemsArray,noOfItems,maxBagWeight)

for i in (populationArray) :
        print(i,'lol')
        
print('--------------------------------------')


print(itemsArray)
for i in range(generations) : 
    mutatedArray  = generateMutatedArray(populationArray,itemsArray,weight,maxBagWeight)
    populationArray = crossOver(populationArray,mutatedArray,crR,maxBagWeight)
    for i in (populationArray) :
        print(getValOfVector(i.vector,itemsArray),i.weight)
        
    print('--------------------------------------')












    


```
